```python
#pyLDAvis for content analysis

from gensim import corpora
import pickle
import gensim
import warnings
import pyLDAvis.gensim
warnings.filterwarnings("ignore", category=DeprecationWarning)
dictionary = gensim.corpora.Dictionary.load('dictionary.gensim')
corpus = pickle.load(open('corpus.pkl', 'rb'))
lda = gensim.models.ldamodel.LdaModel.load('model5.gensim')
lda_data = pyLDAvis.gensim.prepare(lda, corpus, dictionary, sort_topics=False)
pyLDAvis.show(lda_data, ip='127.0.0.1', port=8888, n_retries=50, local=True, open_browser=True, http_server=None)
#pyLDAvis.display(lda_data)
#pyLDAvis.save_html(lda_data, 'test.html')
```
